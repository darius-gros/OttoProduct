En regardant le fichier d'origine, on s'aperçoit que notre matrice de features est "sparse" (clairsemé), la plupart des features sont composés de valeur nulle. Nous devrons en tenir compte dans le choix de notre modèle par la suite.

```python
def percent_nullrows(df = []):
    total = df.count()
    total_null = (df == 0).sum(axis=0)
    return (total_null/total*100)


percent_nullrows(X_train)
```

```python
import seaborn as sns

graph_labels = sns.countplot(y_train)
graph_labels
```

On s'aperçoit que nos classes ne sont pas uniformes, certaines sont plus présentes. De plus si l'on regarde les données on s'aperçoit aussi qu'elles ont été mises en "ordre" par classes les x premières données sont de la classe 1, puis les x données suivantes de la classe 2 et ainsi de suite. Lorsqu'on va séparer nos données d'entraînement pour faire une validation croisée pour tester notre modèle sans toucher à nos odnnées test, il faudra faire attention à spécifier dans le train_test_split shuffle = True pour qu'il y ait une répartition similaire des classes entre notre futur échantillon train et notre échantillon de validation.
Pour la 1ere version de notre modèle nous utiliserons juste un train_test_split.
D'autres personnes ont utilisé des méthodes de sampling pour rééquilibrer les classes. Nous comparerons l'efficacité de notre modèle sans et avec resampling.

```python
from sklearn.preprocessing import LabelEncoder

encod = LabelEncoder()

y_train_labeled = encod.fit_transform(y_train)
y_train_labeled
#c'est une array on va la reshaper en vecteur colonne
y_train_labeled = y_train_labeled.reshape (-1,1)
```

On a utilisé notre label encoder pour transformer les valeurs de notre variable y en valeur numérique discrète pour que le format de valeur soit le bon pour appliquer un modèle et évaluer des métriques sur celui-ci dans un second temps

```python
from sklearn.model_selection import train_test_split

X_train2, X_cv, y_train2, y_cv = train_test_split(X_train, y_train_labeled, test_size = 1/3, shuffle= True)
```

```python
X_train2.head()
```

```python
y_train2
```

Nous allons commencer par utiliser un modèle RandomForest avec les paramètres par défaut. C'est un modèle que nous connaissons et qui a de bonnes performances en général sur les problèmes de classification.
Ce modèles est un modèle qui utilise plusieurs arbres de décision qui vont prédire la variable y sur nos features. Dans une forêt aléatoire, chaque arbre aura un échantillon de données légèrement différent d'un autre arbre ce qui va permettre d'avoir des prédictions légèrement différentes pour chaque arbre. Puis, un vote pondéré des arbres se fera pour décider quelle valeur de y assigner à telle observation. La forêt aléatoire permet de contourner le problème d'overfit des arbres de décisions mais est très gourmande en cpu.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

forest = RandomForestClassifier()
forest.fit(X_train2, y_train2)
pred_y = forest.predict(X_cv)
forest_accuracy = accuracy_score(y_cv, pred_y)
print("accuracy of our random forest model with default parameters is: ", forest_accuracy )

```

Notre accuracy sur notre validation croisée avec notre premier modèle nous donne presque 80% de bonnes prédictions

```python
import matplotlib.pyplot as plt     

cm = confusion_matrix(y_cv, pred_y)

ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax, cmap="Blues");

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 

```
